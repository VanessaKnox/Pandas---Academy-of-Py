```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Change data type of budget to numberic 
school_data["budget"] = pd.to_numeric(school_data["budget"])

#Delete unwanted columns from School Data
del school_data['School ID']
school_data = school_data.rename(columns={"size": "Total_Students"})

student_data = pd.read_csv(student_data_to_load)

#delete uwanted columns from Student data
del student_data['Student ID']


# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#view school_data to see what data is in file
school_data.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Breakdown data in school_data_complete dataframe to extract information needed

#create an array of school names and count number of schools
Total_schools = len(school_data_complete.school_name.unique())

#Count the number of students in data
Total_students = school_data_complete['student_name'].count()

#Determine total budget per school and sum total
Total_budget = school_data['budget'].sum() 

#Calculate average math score
Average_Math_Score = school_data_complete['math_score'].mean()

#Calculate average reading score
Average_Reading_Score = school_data_complete['reading_score'].mean()

#Calculate overall average score (avg math + avg reading)/2 per instructions
Overall_Passing_Rate = ((Average_Math_Score) + (Average_Reading_Score))/2


#Determine if student is passing math and add column to merged dataframe and calc % passing
school_data_complete.loc[school_data_complete.math_score >=70, 'Passing_Math'] = 'Passing'
Passing_math = school_data_complete['Passing_Math'].count()/len(school_data_complete['Passing_Math'])*100


#Determine if student is passing reading and add column to merged dataframe and calc % passing
school_data_complete.loc[school_data_complete.reading_score >=70, 'Passing_Reading'] = 'Passing'
Passing_reading = school_data_complete['Passing_Reading'].count()/len(school_data_complete['Passing_Reading'])*100
```

```python
#Create a district summary data frame from calculations above to hold the results 

District_Summary = {
    "Total Schools":[Total_schools], #return the length of the list
    "Total Students": '{:,}'.format(Total_students),#get the value of the variable and format
    "Total Budget": '${:,.2f}'.format(Total_budget),#get the value of the variable and format
    "Average Math Score": [Average_Math_Score],#get the value of the variable
    "Average Reading Score": [Average_Reading_Score],#get the value of the variable
    "% Students Passing Math": [Passing_math],#get the value of the variable
    "% Students Passing Reading": [Passing_reading],#get the value of the variable
    "Overall Passing Score": [Overall_Passing_Rate]#get the value of the variable
    }

#Convert dict into dataframe
Dist_sum_df = pd.DataFrame(District_Summary)
Dist_sum_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Summarize data by school and then sort by passing rate

#group by school
by_school = school_data_complete.groupby(['school_name'])

#group by type, use.first() or returns a long list of types but only need 1 type per school 
by_type = by_school['type'].first()

#total students by school, .
Student_per_school = by_school.size()

#school budget, use.first() or returns a long list of types but only need 1 budget per school 
sch_budget = by_school['budget'].first()


#Calculate per student budget
stu_budget = sch_budget/Student_per_school


#Calculate Average math score, data coming from Df so can use .mean()
Average_Math_score_school = by_school['math_score'].mean()

#Average reading score
Average_Reading_score_school = by_school['reading_score'].mean()

#Average passing math score, count the number of students passing math 
#('Passing_Math' calculated in district summary and appended to the table). *100 to get %
pass_math = school_data_complete.groupby('school_name')['Passing_Math'].count()/Student_per_school*100

#Average passing reading score, count the number of students passing math 
#('Passing_Reading' calculated in district summary and appended to the table). *100 to get %
pass_reading = school_data_complete.groupby('school_name')['Passing_Reading'].count()/Student_per_school*100

#Overall Passing score (reading+math)/2 per instructions
over_all_pass = (pass_math + pass_reading )/2

#create a dictionary that hold the results from above, don't use[] around variable or will create a list and values are needed to format
School = {
    'School Type': by_type,
    'Total Students': Student_per_school, 
    'Total Budget': sch_budget,
    'Per Student Budget': stu_budget,
    'Average Math Score': Average_Math_score_school,
    'Average Reading Score': Average_Reading_score_school,
    '% Students Passing Math': pass_math,
    '% Students Passing Reading': pass_reading,
    'Overall Passing Score': over_all_pass
       }
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#create panda dataframe from dictionary above and format
School_Summary = pd.DataFrame(School)
formatted_school_summary = School_Summary.copy()
formatted_school_summary['Total Budget'] = formatted_school_summary['Total Budget'].map('${:,.2f}'.format)
formatted_school_summary['Per Student Budget'] = formatted_school_summary['Per Student Budget'].map('${:,.2f}'.format)
formatted_school_summary.head()
```

```python
#Sort dataframe by top 5 Overall Passing Score, ascending=False will sort in highest to lowest.  descending sort is assumed so if want ascending dictate

Top_5_Schools = formatted_school_summary.sort_values(by= 'Overall Passing Score',ascending=False)
Top_5_Schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
#Sort dataframe by top 5 Overall Passing Score, ascending=False will sort in highest to lowest.  descending sort is assumed so if want ascending dictate

Bottom_5_Schools = formatted_school_summary.sort_values(by= 'Overall Passing Score',ascending=True)
Bottom_5_Schools.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Calculate the average math score by grade 9 per school
Avg_Math_School_9 = school_data_complete.loc[school_data_complete.grade == '9th'].groupby('school_name')['math_score'].mean()

#Calculate the average math score by grade 10 per school
Avg_Math_School_10 = school_data_complete.loc[school_data_complete.grade == '10th'].groupby('school_name')['math_score'].mean()

#Calculate the average math score by grade 11 per school
Avg_Math_School_11 = school_data_complete.loc[school_data_complete.grade == '11th'].groupby('school_name')['math_score'].mean()

#Calculate the average math score by grade 12 per school
Avg_Math_School_12 = school_data_complete.loc[school_data_complete.grade == '12th'].groupby('school_name')['math_score'].mean()

#Create a dictionary to hold values:
Avg_Math = {'9th Grade': Avg_Math_School_9,
         '10th Grade': Avg_Math_School_10,
           '11th Grade': Avg_Math_School_11,
           '12th Grade': Avg_Math_School_12}

#create dataframe to display values in the dictionary
Avg_Math_School_Grade = pd.DataFrame(Avg_Math)
Avg_Math_School_Grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins
spending_bins = [0, 584.99, 614.99, 644.99, 999999]
group_names = ["<$585", "$585-614", "$615-644", "> $644"]
```

```python
#Create a new data frame with columns matching criteria
Scores_by_Spending = School_Summary.loc[:,['Average Math Score',
                                           'Average Reading Score',
                                          '% Students Passing Math',
                                          '% Students Passing Reading',
                                          'Overall Passing Score',]]

#Add a new column called Spending Ranges= then bin based off budget per student
Scores_by_Spending['Spending Ranges (Per Student)'] = pd.cut(School_Summary['Per Student Budget'], spending_bins,labels=group_names)

#display dataframe using ('Spending Ranges (Per Student)') as column header
Scores_by_Spending = Scores_by_Spending.groupby('Spending Ranges (Per Student)').mean()

Scores_by_Spending.head()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins
size_bins = [0, 999, 1999, 99999999]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large >2000"]
```

```python
#Create a new data frame with columns matching criteria
Scores_by_Size = School_Summary.loc[:,['Average Math Score',
                                           'Average Reading Score',
                                          '% Students Passing Math',
                                          '% Students Passing Reading',
                                          'Overall Passing Score',]]

#Add a new column called 'Average Scores (Per School Size)' = then bin based off 'Total students'
Scores_by_Size['Average Scores (Per School Size)'] = pd.cut(School_Summary['Total Students'], size_bins, labels=group_names)

Scores_by_Size = Scores_by_Size.groupby('Average Scores (Per School Size)').mean()

Scores_by_Size
```
